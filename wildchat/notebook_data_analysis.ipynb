```python
import re
import json
import pickle

def load_pickle_file(file_path):
    try:
        # Open the file in read-binary mode ('rb')
        with open(file_path, 'rb') as file:
            data = pickle.load(file)
        return data
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return None
    except pickle.UnpicklingError as e:
        print(f"Error unpickling data: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

def load_json_file(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as file:
            data = json.load(file)
        return data
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return None
    except json.JSONDecodeError:
        print(f"Error: Could not decode JSON from the file '{filepath}'. Check for valid JSON syntax.")
        return None

def read_prompt_from_text_file(path_to_text_file):
    with open(path_to_text_file, 'r') as file:
        prompt_str = file.read()
    return prompt_str
```

```python
output_dir= "/home/kramesh3/projects/dynamic_policy/wildchat/wildchat_queries"
prompt_style = "style2"
prompt_file_path = f"/home/kramesh3/projects/dynamic_policy/wildchat/prompts/{prompt_style}.txt"
file_name = f"privacy_risks-{prompt_style}-version-main-0-10000.json"

pkl_file = load_json_file(f"{output_dir}/{file_name}")
prompt_text = read_prompt_from_text_file(prompt_file_path)
```

```python
import pandas as pd

# Prepare data for CSV
data = []

for i in pkl_file:
    if re.search(r'privacy\s*.*?\s*risk\s*.*?\s*level\s*.*?\s*yes', i['model_answer'], re.IGNORECASE):
        prompt = i['full_text'].replace(prompt_text[:-183], "").replace("system\nYou are an expert judge.\nuser\n\n", "").replace(prompt_text[-160:], "").replace(i['model_answer'], "")
        
        data.append({
            'PROMPT': prompt,
            'MODEL_ANSWER': i['model_answer']
        })

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv('results.csv', index=False, encoding='utf-8')

print(f"CSV file 'results.csv' created successfully with {len(df)} rows")
print(df)  # Display the dataframe
```
